## Conclusion
The univariate, bivariate, and multivariate analyses provides valuable insights into passenger behavior, transport mode efficiency, fare patterns, and temporal trends in trip data.

From the univariate analysis, we observed that variables such as Passenger_Count, Fare_Amount, and Trip_Duration_Minutes exhibit varying degrees of skewness and dispersion. Passenger and fare distributions are positively skewed, with most values concentrated below 50, while trip durations displayed high variability with a near-normal distribution. This highlights the presence of a diverse passenger base and a somewhat standardized, zone-based fare system, but varying travel durations likely influenced by traffic, distance, or route type.

The bivariate analysis revealed meaningful associations. Count plots showed that buses are the most used mode of transport, with ferries following and trams being the least utilized. Time-based trends indicated that trip activity peaks on Sundays and Saturdays, pointing to increased mobility during weekends. A closer look at transport mode efficiency indicated that trams have the shortest and most consistent trip durations, making them the most time-efficient, while ferries exhibit the greatest variability. Average fare amounts for buses, trams, and trains were relatively uniform, with ferries deviating from this pattern.

The multivariate analysis, including heatmaps and line plots, revealed temporal dependencies and usage patterns across days and hours. For instance, train services peaked in the early morning (8:00 AM), buses in the early afternoon (1:00 PM), and trams in the late afternoon (4:00 PM), reflecting distinct commuting patterns across transport types. Passenger counts peaked at 8:00 AM and 9:00 PM, likely aligned with morning and evening commute hours, while weekend usage clearly surpassed weekday levels, particularly in terms of overall passenger count and trip activity.
